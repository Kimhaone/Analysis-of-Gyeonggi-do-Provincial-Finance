```python
import pandas as pd
import re
import time
```

```python
#code1. 기본데이터 열기-필요한 자료들 분리
pd_person = pd.read_table('GGWFF_DATA_FINAL.txt', sep=',', header=None,
                          names = ['ID', '생년', '성별', '우편번호', '지역', '직업'], encoding = 'utf=16')
#pd_person = pd_person[['ID', '생년', '성별', '지역']]
pd_person.head()
```

```python
#code3. 칼럼 헤더 열기
header_df = pd.read_excel('header.xlsx', sheet_name = 'Sheet3')
header_names = header_df.columns.tolist()
header_names
```

```python
#code4. 대출건수 불러오기
start = time.time()
pd_id = pd.DataFrame(columns = ['ID',  '신용카드이용금액'])

chunksize = 10**4
for cnt, chunk in enumerate(pd.read_csv('GGWFF_DATA_202111_JOIN.txt', names =  header_names, header = None, chunksize =chunksize)):
    temp = chunk[['ID','신용카드이용금액']]

    #print(temp.head())
    pd_id = pd_id.append(temp)
#    print(cnt)
#    if cnt >= 1400:
#        break
    if chunk is None: 
        break
pd_id.head()
print(time.time()-start)
```

```python
pd_person = pd_person.sort_values(by = ['ID'], axis = 0)
pd_id = pd_id.sort_values(by = ['ID'], axis = 0)
pd_person.head()
pd_id.head()
```

```python
pd_person.insert(1, "연령대", (2023-pd_person['생년'])/10)
pd_person['연령대'] = pd_person['연령대'].astype(int)

pd_person.insert(2, "연령", (2023-pd_person['생년']))
pd_person['연령'] = pd_person['연령'].astype(int)

pd_person.head()
```

```python
pd_person['지역'] = pd_person['지역'].str.extract(r'([가-힣]+)')
```

```python
pd_rperson = pd_person[['ID', '연령대', '연령', '성별', '지역', '직업']]
```

```python
pd_result = pd.merge(pd_person, pd_id, how = 'outer', on = 'ID')
pd_result.head()
```

```python
#경기도민의 총카드 개설건수 구하기
pd_total=pd_result.query('신용카드이용금액 > 0')
pd_total20=pd_result.query('신용카드이용금액 > 0 and 20<=연령<25')
pd_total25=pd_result.query('신용카드이용금액 > 0 and 25<=연령<30')
pd_total30=pd_result.query('신용카드이용금액 > 0 and 30<=연령<35')
pd_total30.head()
```

```python
debit_total = pd_total['신용카드이용금액'].sum()
debit_mean = pd_total['신용카드이용금액'].mean()
num_ct = len(pd_total.index)

print('신용카드이용금액:', debit_total, '평균신용카드이용금액:', debit_mean, '인원수: ', num_ct)
```

```python
#남성인경우
pd_m_total = pd_total30.query('성별 == "M"')
debit_m_total = pd_m_total['신용카드이용금액'].sum()
debit_m_mean = pd_m_total['신용카드이용금액'].mean()
num_m_ct = len(pd_m_total.index)

#여성만 추출
pd_w_total = pd_total30.query('성별 == "W"')
debit_w_total = pd_w_total['신용카드이용금액'].sum()
debit_w_mean = pd_w_total['신용카드이용금액'].mean()
num_w_ct = len(pd_w_total.index)

print('남성: 신용카드이용금액:', debit_m_total, '평균신용카드이용금액:', debit_m_mean, '인원수: ', num_m_ct)
print('여성: 신용카드이용금액:', debit_w_total, '평균신용카드이용금액:', debit_w_mean, '인원수: ', num_w_ct)
```
